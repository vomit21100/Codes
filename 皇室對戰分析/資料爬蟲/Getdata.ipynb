```python
"""
靜態爬蟲
requests.get會拿到網頁設計和包含玩家資料的json字串
"""
response = requests.get('https://royaleapi.com/players/leaderboard', headers = headers)
soup = bs(response.text, "html.parser")

data = soup.select('script')

data = data[15]
text = data.get_text()
start = text.find('[')-1
end = text.rfind(']')+1
```

```python
#load->讀取檔案 loads->字串轉換
JsonData = json.loads(text[start : end])
```

```python
#前1000名玩家tag
for i in JsonData:
    print(i['tag'])
```

```python
"""
好像不需要動態爬蟲...
"""

import random
import os
from datetime import date

#today = date.today().strftime("%Y-%m-%d")
folder = '../data/' + date.today().strftime("%Y-%m-%d") + '/'
if not os.path.exists(folder):
    os.mkdir(folder)
    
index = 0
for i in JsonData:
    print(index, i['tag'])
    url = 'https://royaleapi.com/player/' + i['tag'] + '/battles/csv'
    data = requests.get(url, headers = headers)
    with open(folder + i['tag'] + '.csv', 'wb') as file:
        file.write(data.content)
        file.close()
    time.sleep(random.randint(30, 40))
    index += 1
    
print('Finish')
```

```python
import pandas as pd
import os

#整理資料 只留天梯 等級不管
wanted = ['type','battleTime',
          'team_0_tag','team_0_crowns',
          'opponent_0_tag','opponent_0_crowns',
          'team_0_cards_0_name','team_0_cards_1_name','team_0_cards_2_name','team_0_cards_3_name','team_0_cards_4_name','team_0_cards_5_name','team_0_cards_6_name','team_0_cards_7_name', 
          'opponent_0_cards_0_name','opponent_0_cards_1_name','opponent_0_cards_2_name','opponent_0_cards_3_name','opponent_0_cards_4_name','opponent_0_cards_5_name','opponent_0_cards_6_name','opponent_0_cards_7_name',]

os.chdir('../data')
folders = os.listdir()
if 'test.csv' in folders:
    folders.remove('test.csv')
#print(files)

df = pd.DataFrame()
for folder in folders:
    #print(folder)
    for i in os.listdir(folder):
        #print(i)
        df1 = pd.read_csv(folder + '/' + i, index_col=False)
        df = pd.concat([df, df1])

    
#保留特定資料
fliter = (df['type'] == 'pathOfLegend')
df = df[fliter]

#留下需要的column
df = df[wanted]

#去掉重複的
df = df.drop_duplicates(subset=['battleTime'])

df.to_csv('../merged.csv', encoding='utf-8', index=False)
print(df.shape[0])
    
os.chdir('../資料爬蟲')
```
